```julia
N=2000;
```

```julia
b = randn(N,N);
```

```julia
for ii=1:N
    for jj=1:N
        b[ii,jj]=(sin(ii+jj^2)+sin(jj+ii^2))/2
    end
end
```

```julia
@time c=eigen(b);
```

```julia
eigenvalues=c.values;
```

```julia
scatter(1:N,c.values)
```
