```julia
using LinearAlgebra
using Plots
#plotlyjs()
```

```julia
Nplotpts = 51;
x_array = range(-1.0,1.0,Nplotpts);
y_array = range(-3.0,3.0,Nplotpts);

f_array = zeros(Nplotpts,Nplotpts);

for ii = 1:Nplotpts
    for jj = 1:Nplotpts
        f_array[ii,jj] = x_array[ii]^2 + y_array[jj];
    end
end
```

## Note: It has to be f_array'. Otherwise, x and y axis would be flipped.

```julia
heatmap(x_array, y_array, f_array', color = :rainbow, xlims=(-1.0,1.0), ylims=(-3.0,3.0), aspect_ratio=:equal, xlabel="x", ylabel="y", title="Plot of f=x^2 + y")
```
