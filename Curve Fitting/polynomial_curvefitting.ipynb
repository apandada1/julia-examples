```julia
using Plots; using LsqFit
```

```julia
N=30;
```

```julia
x=range(0,1,N);
```

```julia
y=x.^2 + (2*rand(N)-ones(N))/20;
```

```julia
@. model_function(x, p) = p[1]+p[2]*x+p[3]*x^2
```

```julia
initial_params = [0.1,0.1,2.0]
```

```julia
plt1 = plot(x,y)
```

```julia
fit_result_nonlinear = curve_fit(model_function, x, y, initial_params)
```

```julia
results_param = fit_result_nonlinear.param
```

```julia
plt1 = plot!(x, model_function(x, results_param))
```
